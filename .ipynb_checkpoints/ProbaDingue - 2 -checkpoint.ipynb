The idea would be that an isolate bar has no effect on his neighours. However together they have an impact on each of them. The Group has an influence on the singular bar.

What is a Group ?

A group is constituted of minimum 2 members. All members of a group must at least share a common feature, otherwise they would not be part of the group.

An important question is : What is the criterion that defines a group ? In other words, what is linking every members of the group ? Usually, the criterion used to judge the link between members is the covariance. However in our case the effect between 2 members may not be the same in both directions. It remains to find a good alternative

What kind of sets can we find ?
Groups may not be present all the time. We may have period of total randomness

# First model : The Trend Group

This first model will try to capture the group with a common trend. Some bars seems to be part of a common trend. In other words, there seems to exists a group where positive returns dominates over negative ones. Sometime negative returns inside those groups are insignificant for the integrity of those group. However even those negative isolated bars are part of the group, they should therefore share something with the group.

This first group comes from my intuition by looking at charts. However I should try to develop a method for finding some other potential groups

The most simple model I could think of when it comes to this group is the following, 


\begin{align*}
    \large
    X_t = \epsilon_t + \alpha G_t && \text{ where } \epsilon_t \sim GWN(0,\sigma^2_\epsilon)
\end{align*}

This models the time series as a white noise with a constant which is not always active. 


There are 2 important parameters in this model : 

* $\alpha \in (-\infty, \infty)$ : which gives the intensity of the trend group. It is first assumed constant.
* $G_t \in \{0,1\}$ : which gives us when the trend group is active. $1$ if the group is active at time $t$.

The effect of the group "trend" on the observation is to shift its expectation positively or negatively. 

The parameter $G_t$ must depend on past and future values since, by definition, the group affects multiple neighbouring observations. We must at least have some kind of clustering of the same values of $G_t$.


The random variable $G_t$ is a Bernoulli r.v. which has a probability $\pi$ of being $1$. However we need that probability to depend on previous probabilities (at least 1) if we want to have that kind of clustering. We then make that probability follow an AR(1) model (could be extended to AR(p)). Note that $\pi_t$ is a probability and that the AR(1) does not provide values between 0 and 1. Therefore we pass that AR(1) model into a sigmoid function $f: \mathbb{R} \rightarrow [0,1]$

\begin{align*}
    G_t \sim Bernoulli(\pi_t)\\
    \pi_{t+1} = f(z_t) && \text{ where } f:= \text{ sigmoid }\\
    z_{t+1} = \omega + \phi z_t + \nu_t && \text{ where } \nu \sim GWN(0, \sigma^2_{\nu})
\end{align*}

```python
def sigmoid(x, lam:np.float64):
    return 1.0 / (1.0 + np.exp(-lam*x))
```

```python
def bernoulli(p):
    return (np.random.uniform(0,1) <= p)*1.0
```

```python
def AR(phi:np.float64, const:np.float64=0.0, var_e:np.float64=1.0):
    previous_value = np.random.normal(0.0, var_e)
    
    while True:
        current_innov = np.random.normal(0.0, var_e)
        current_value = const + phi * previous_value + current_innov
        yield current_value
        previous_value = current_value
```

```python
def G_process(phi:np.float64, const:np.float64=0.0, var_e:np.float64=1.0, lam:np.float64=1.0):
    ar_process = AR(phi, const, var_e)
    
    while True:
        current_pi = sigmoid(next(ar_process), lam)
        yield bernoulli(current_pi)
```

```python
def trend_process(alpha:np.float64, var_e:np.float64, lam:np.float64, const:np.float64, phi:np.float64):
    g_process = G_process(phi, const, var_e, lam)
    while True:
        yield np.random.normal(0,1) + alpha * next(g_process)
```

```python
def sample_process(process, n:int, *args):
    sample = np.empty((n, ), dtype=np.float64)
    process = process(*args)
    
    for i in range(n):
        sample[i] = next(process)
    
    return sample
```

```python
# Get the max. likelihood estimator only for AR(1) model
def maximum_likelihood_function(params:np.ndarray, sample:np.ndarray):
    likelihood = (sample[1:] - params[1] - params[0] * sample[:-1])**2
    likelihood = np.sum(likelihood)
    return -(-len(sample)/2.0 * np.log(2*np.pi*params[2]) - likelihood/(2*params[2]))
```

## Model Trend simply

The previous model was too complicated and I will return to it later. I believe simplicity is key to trading. In the sense that I need simple ideas in the first place and then I make things more complicated after I have understood the simple model and where it needs improvements. 

The model I build know has the following form, 
\begin{align}
    \large
    X_t = \epsilon_t + \alpha \mathbb{1}_{A_t}
\end{align}
where $A_t$ is the event such that : $\epsilon_{t-1} \geq \bar{\alpha}$.

This characterization models the time series $X_t$ as random where there is occasionally an increased constant. In order to induce a Trend, I need that "occasionally" condition to depend on previous observations, e.g. on the lag 1 white noise for the most simple case.

We can check if this model is weakly-stationary by checking its moments. Let $\mathbb{P}(\epsilon_{t-1} \geq \bar{\alpha}) = p$: 
\begin{align}
    \mathbb{E}(X_t) &= \alpha p \\
    \mathbb{C}ov(X_t, X_{t-h}) &= 
                    \begin{cases}
                        \sigma_\epsilon^2 + \alpha^2 p (1-p) &\text{ if } h=0\\
                        \alpha p \mathbb{E}(\epsilon_t | \epsilon_{t} \geq \bar{\alpha}) &\text{ if } h=1\\
                        0 &\text{ if } h > 1
                    \end{cases}
\end{align}

Note that eventhough we have the truncated expectation, the latter does not depend on time, only on the threshold ! 

We can say that this process is then weakly-stationary.

Note also that we have some overall correlation between observations $X_t$. This dependence is however not always active ! we need some conditions, namely the conditions of activations of the indicator function.


```python
def trend_process(strength:np.float64, proba:np.float64):
    previous_innov = np.random.normal(0, 1)
    thresh = sc.norm.ppf(1.0 - proba, 0, 1)
    
    while True:
        current_innov = np.random.normal(0,1)
        indicator = 1.0 if previous_innov >= thresh else 0.0
        yield current_innov + strength * indicator
        previous_innov = current_innov
```

```python
strength = 0.25
p = 0.1
thresh = sc.norm.ppf(1.0 - p)
sample_trend = sample_process(trend_process, 20000, strength, p)

print('mean : ', strength*p, '   ', np.mean(sample_trend))
print('var : ', 1.0+(strength**2)*p*(1-p),'   ', np.var(sample_trend))

trunc_mean = sc.norm.pdf(thresh) / (1.0 - sc.norm.cdf(thresh))
empirical_lag1 = np.cov(sample_trend[1:], sample_trend[:-1])[0,1]
print('autocov. lag1 : ',trunc_mean*p*strength, '   ', empirical_lag1)
```

```python
# params : sigma_e(0), strength(1), thresh(2)
def cond_normal_1(params:np.ndarray, prev_x:np.float64):
    var_e = params[0]**2
    strength = params[1]
    p = 1.0 - sc.norm.cdf(params[2], 0, np.sqrt(var_e))
    
    cond_mean = (var_e*(prev_x - strength*p))/(var_e + strength*strength*p*(1-p)) 
    cond_var = (var_e) * (1.0 - (var_e/(var_e + strength*strength*p*(1-p))))
    
    return 1.0 - sc.norm.cdf(params[2], cond_mean, np.sqrt(cond_var))
```

```python
def cond_expectation_1(params:np.ndarray, prev_x:np.float64):
    return params[1]*cond_normal_1(params, prev_x)
```

```python
def cond_var_1(params:np.ndarray, prev_x:np.float64):
    c_norm = cond_normal_1(params, prev_x)
    return (params[0]**2) + (params[1]**2) * c_norm*(1-c_norm)
```

```python
def log_likelihood2(params:np.ndarray, sample:np.ndarray):
    prev_x = sample[:-1]
    x = sample[1:] - cond_expectation_1(params, prev_x)
    var = cond_var_1(params, prev_x)
    
    likeli = np.log(var)/2.0 + (x**2)/(2.0*var)
    
    return np.sum(likeli)
```

```python
def constr(params:np.ndarray):
    var_e = params[0]**2
    strength = params[1]
    p = 1.0 - sc.norm.cdf(params[2], 0, np.sqrt(var_e))
    return strength*strength*p*(1-p)
```

```python
x0 = [1.0, 0.1, 2.0]
const = ({'type': 'ineq', 'fun': constr})

bnds = ((0.0, None), (-10, 10), (-10, 10))

res = minimize(log_likelihood2, x0, constraints=const, bounds=bnds, args=(np.array(eurusd)))
res.x
```
